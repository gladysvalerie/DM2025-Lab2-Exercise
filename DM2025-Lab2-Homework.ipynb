## 1. Model Development (10 pts Required)

### 1.1 Preprocessing Steps
I loaded the three data: `data_identification.csv`, `emotion.csv`, and `final_posts.json`. Then I put the csv files to two dataframes and I need to flatten the JSON file to extract `post_id` and `text`. Then I merge all of them to a single dataframe `df`.

- Lowercasing (for normalization)
- URL removal (not semantically meaningful for emotion)
- Mention removal (remove @username thing, if exists)
- Hashtag symbol removal
- Keeping only letters and emotional punctuation
- Whitespace normalization
  
Then I had to scan for empty texts and duplicate texts, normal stuff.

### 1.2 Feature Engineering Steps
I used OpenAI's `text-embedding-3-large` model to generate the embeddings, with 3072-dimensional vectors. I used batch processing with the batch size 1000, so it's quite efficient. The larger embedding space allows for richer feature representation, so I thought that it may be the best embedding model. Additionally, the API-based approach simplified the pipeline by eliminating the need for local GPU resources or model fine-tuning.

### 1.3 Model Explanation
I used CatBoost classifier (gradient boosting), mainly because it allows for GPU training and it handles the class imbalance. It was because I tried using XGBoost and it took so long and give the somewhat similar result. These are the hyperparameters that I used.
`iterations=1000`
`depth=8`
`learning_rate=0.05`
`loss_function='MultiClass'`
`auto_class_weights='Balanced' â€” explain this handles class imbalance`
`task_type='GPU'`

I used 5-fold stratified cross-validation with the metric macro f-1 score. I got 0.4880 for the CV score.

```python
from google.colab import userdata
from openai import OpenAI
import tqdm

api_key = userdata.get("OPENAI_API_KEY")
client = OpenAI(api_key=api_key)
```

```python
import pandas as pd
import json
import re

# Load files
id_df = pd.read_csv(BASE_DIR + "/data_identification.csv")
emo_df = pd.read_csv(BASE_DIR + "/emotion.csv")

with open(BASE_DIR + "/final_posts.json") as f:
    raw_posts = json.load(f)

# Flatten JSON structure
posts = []
for item in raw_posts:
    src = item["root"]["_source"]["post"]
    posts.append({
        "id": src["post_id"],
        "text": src["text"]
    })

post_df = pd.DataFrame(posts)

# Merge
df = id_df.merge(post_df, on="id", how="left")
df = df.merge(emo_df, on="id", how="left")

df.head()
```

```python
def clean_text(text):
    if pd.isna(text):
        return ""
    text = str(text).lower()
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)  # remove URLs
    text = re.sub(r'@\w+', '', text)  # remove mentions
    text = re.sub(r'#', '', text)  # remove hashtag symbol
    text = re.sub(r'[^a-zA-Z\s!?]', ' ', text)  # keep letters and !?
    text = re.sub(r'\s+', ' ', text).strip()  # remove extra spaces
    return text

df['clean_text'] = df['text'].apply(clean_text)
df[['text', 'clean_text']].head()
```

```python
train_df = df[df["split"]=="train"]
test_df  = df[df["split"]=="test"]
```

```python
# Check for empty strings or NaN
train_texts = train_df['clean_text'].tolist()

empty_count = sum(1 for t in train_texts if t == "" or t is None or len(t.strip()) == 0)
print(f"Empty texts: {empty_count}")

# Find examples of problematic texts
for i, t in enumerate(train_texts[:1000]):
    if t == "" or t is None or len(t.strip()) == 0:
        print(f"Index {i}: '{t}'")
```

```python
import numpy as np

def get_openai_embeddings(texts, batch_size=1000):
    # Replace empty strings with a placeholder
    texts = [t if t and len(t.strip()) > 0 else "empty" for t in texts]

    embeddings = []
    total = len(texts)
    for i in range(0, total, batch_size):
        batch = texts[i:i+batch_size]
        response = client.embeddings.create(
            model="text-embedding-3-large",
            input=batch
        )
        batch_emb = [item.embedding for item in response.data]
        embeddings.extend(batch_emb)
        print(f"Progress: {min(i+batch_size, total)}/{total}")
    return np.array(embeddings)

print("Encoding train...")
X_train = get_openai_embeddings(train_df['clean_text'].tolist())

print("\nEncoding test...")
X_test = get_openai_embeddings(test_df['clean_text'].tolist())

print(f"\nDone! Shape: {X_train.shape}")
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
y_train = le.fit_transform(train_df['emotion'])
print("Classes:", le.classes_)
```

```python
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold

# No SMOTE, just class weights
model = CatBoostClassifier(
    iterations=1000,
    depth=8,
    learning_rate=0.05,
    loss_function='MultiClass',
    auto_class_weights='Balanced',
    random_seed=42,
    verbose=100,
    task_type='GPU'
)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='f1_macro')
print(f"CV F1 (No SMOTE): {scores.mean():.4f}")
```

```python
model.fit(X_train, y_train)

# Predict
y_pred = model.predict(X_test)
y_pred_labels = le.inverse_transform(y_pred.flatten())

# Create submission
submission = pd.DataFrame({
    'id': test_df['id'],
    'emotion': y_pred_labels
})

submission.to_csv('submission_cat1.csv', index=False)
print("Saved!")
submission.head(10)
```

```python
from transformers import pipeline

classifier = pipeline(
    "text-classification",
    model="j-hartmann/emotion-english-distilroberta-base",
    top_k=1,
    device=0  # GPU
)

# Test on one sample
test_text = test_df['clean_text'].iloc[0]
result = classifier(test_text)
print(f"Text: {test_text}")
print(f"Result: {result}")
```

```python
# This model outputs: anger, disgust, fear, joy, neutral, sadness, surprise
# Your competition has: anger, disgust, fear, joy, sadness, surprise (no neutral)

# Let's see what labels it gives
sample_results = classifier(test_df['clean_text'].head(10).tolist())
for text, res in zip(test_df['clean_text'].head(10), sample_results):
    print(f"{res[0]['label']:10} | {text[:50]}...")
```

```python
from tqdm import tqdm

def predict_batch(texts, batch_size=64):
    predictions = []
    for i in tqdm(range(0, len(texts), batch_size)):
        batch = texts[i:i+batch_size]
        results = classifier(batch)
        for res in results:
            label = res[0]['label']
            # Handle 'neutral' - map to most common class or closest
            if label == 'neutral':
                label = 'joy'  # or we can use our CatBoost for these
            predictions.append(label)
    return predictions

test_texts = test_df['clean_text'].tolist()
preds_pretrained = predict_batch(test_texts)
print("Done!")
```

```python
# Test on training data to see F1
train_texts = train_df['clean_text'].tolist()
preds_train = predict_batch(train_texts)

from sklearn.metrics import f1_score
f1 = f1_score(train_df['emotion'], preds_train, average='macro')
print(f"Pre-trained model F1 on train: {f1:.4f}")
```

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from lightgbm import LGBMClassifier
from catboost import CatBoostClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold

ensemble = VotingClassifier(
    estimators=[
        ('lr', LogisticRegression(max_iter=1000, class_weight='balanced', random_state=42)),
        ('lgbm', LGBMClassifier(n_estimators=500, class_weight='balanced', random_state=42, verbose=-1)),
        ('cat', CatBoostClassifier(iterations=500, auto_class_weights='Balanced', random_seed=42, verbose=0)),
    ],
    voting='hard'
)

cv = StratifiedKFold(n_splits=3, shuffle=True, random_state=42)
scores = cross_val_score(ensemble, X_train, y_train, cv=cv, scoring='f1_macro')
print(f"Ensemble F1: {scores.mean():.4f}")
```

```python
ensemble.fit(X_train, y_train)
preds_ensemble = ensemble.predict(X_test)
preds_ensemble_labels = le.inverse_transform(preds_ensemble)
print("Ensemble predictions done!")
```

```python
# If both models agree â†’ use that
# If they disagree â†’ use the one with higher confidence (ensemble)

final_predictions = []
for i in range(len(test_df)):
    if preds_pretrained[i] == preds_ensemble_labels[i]:
        final_predictions.append(preds_pretrained[i])
    else:
        # Use ensemble when they disagree (usually more reliable with good embeddings)
        final_predictions.append(preds_ensemble_labels[i])

print("Combined predictions done!")
```

```python
submission = pd.DataFrame({
    'id': test_df['id'],
    'emotion': final_predictions
})

submission.to_csv('submission_combined.csv', index=False)
print("Saved!")
submission['emotion'].value_counts()
```
